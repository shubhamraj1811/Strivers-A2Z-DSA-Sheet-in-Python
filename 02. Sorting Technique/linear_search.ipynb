# 🔥 Linear Search

## Concept
- Linear search is exactly what it sounds like: 
- Scanning an array in a straight line. 
- Imagine looking for a specific book on a completely disorganized shelf. 
- You have no shortcuts; you must check the first book, then the second, then the third, until you find your target or run out of books.

## Pythonic Implementation
- In interviews, do not use clumsy while loops with manual index counters for a simple linear search.
- Use Python's built-in enumerate to keep your code clean and readable.

```python
def linear_search(arr, target):
    # Returns the index of the target if found, otherwise -1.
    for i, val in enumerate(arr):
        if val == target:
            return i  # Target found, exit immediately
    return -1  # Target not in the array

# main
my_list = [1,2,3,4,5]
print(linear_search(my_list, 5))
print(linear_search(my_list, 9))
```

## Complexity Analysis

### Time Complexity

**Worst Case:** $O(N)$ - the target is at the very end, or doesn't exist at all.
**Best Case Time:** $O(1)$ - you get incredibly lucky and the target is the very first element

### Space Complexity

**Worst/Best Case:** $O(1)$ - we are only storing a single variable i in memory to track our position; no extra arrays are created.
