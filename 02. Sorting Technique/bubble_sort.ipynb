# 🔥 BUBBLE SORT

## Concept of Bubble Sort
- Bubble sort is an sorting algorithm, where we compare two adjacent element and if one is larger than the other, we move it ahead.
- We can think of bubble sort as a process of repeatedly fixing local disorder

### Conclusion
- An important conclusion from bubble sort is that after the first pass, the largest elements comes to the right side.
- After each pass, the sorting area is reduced by -1
- Since the most recent last position are already sorted

## Analogy
- Suppose there is a line of students with different heights.
- A queue of students arranged by height.
  - If left student is taller than right student → swap them
  - Else → keep them
- After walking through the whole line once, the tallest student reaches the end.


## Example
**Array to be sorted = [3,5,11,7,2]**

- compare 3 and 5 -> no swap is needed
- compare 5 and 11 -> no swap is needed
- compare 11 and 7 -> swap
- array becomes = [3,5,7,11,2]
- compare 11 and 2 -> swap
- array = [3,5,7,2,11]

We can see that after the first pass : largest element (11) was on the right side

```python
# python implementation


def bubble_sort(arr):
    n = len(arr)

    for i in range(n - 1):  # passes
        for j in range(n - 1 - i):  # comparisons
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

    return arr

# main
arr = [3, 5, 11, 7, 2]
print(f"Unsorted Array: {arr}")

bubble_sort(arr)
print(f"After Bubblesort: {arr}")
```

## Time Complexity
**Worst Case: O(n²)**
**Best Case: O(n²)**

## Optimized Version of Bubble Sort
- We can stop early if no swaps happen in the first pass
- Why? If a complete pass makes 0 swaps, then the array is already sorted

```python
# optimized version

def bubble_sort(arr):
    n = len(arr)

    for i in range(n - 1):
        swapped = False

        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True

        if not swapped:
            break

    return arr

# main
arr = [3, 5, 11, 17, 22]
print(f"Unsorted Array: {arr}")

bubble_sort(arr)
print(f"After Bubblesort: {arr}")
```

## Time Complexity
Now Best case becomes **O(N)**
