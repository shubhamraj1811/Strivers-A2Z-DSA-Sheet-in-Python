# Dictionary

## What is Dictionary
- A dictionary is just an array that lets you choose your own indices. 
- Under the hood, Python dictionaries (dict) use the exact same hash table mechanics as sets. 
- But while sets only store a key to check for existence, a dictionary stores a Key-Value pair. 
- This makes it the ultimate tool for mapping relationships, counting frequencies, and caching data (Memoization) in dynamic programming.

### Dictionary in Interviews
- Because dictionaries use a hash table, looking up a value by its key, inserting a new pair, or deleting a pair all take $O(1)$ average time. 
- In a FAANG interview, your first instinct when you see an $O(N^2)$ brute-force solution with nested loops should almost always be: "Can I use a dictionary to trade $O(N)$ space for $O(1)$ lookup time?"

## Creating Dictionaries
You can initialize them empty or pre-populated.

```python
# Empty dictionary
empty_dict = {}
# or
hash_map = dict()

print(empty_dict)
print(type(empty_dict))

# Pre-populated
user_profile = {
   "name": "Shubham", 
   "role": "Android Developer", 
   "language": "Kotlin"}

print(user_profile)
```

## Accessing Elements (The Interview Trap)
- There are two ways to get a value out of a dictionary.
- Interviewers will watch closely to see which one you use.

Pro Tip :- Always use .get() in interviews unless you have already mathematically proven the key exists in previous logic.

```python
# Method 1: Bracket notation
name = user_profile["name"]
print(name)
# TRAP: user_profile["age"] throws a KeyError and crashes your program.

# Method 2: The .get() method (Safer)
age = user_profile.get("age")  # Returns None instead of crashing
age = user_profile.get("age", 25)  # Returns 25 as a default fallback if "age" doesn't exist
```

## Modifying Elements
- Updating a dictionary is straightforward. 
- If the key exists, it overwrites the value. 
- If it doesn't exist, it creates a new key-value pair.
- Both are $O(1)$ operations.

```python
dict2 = {
   "name": "Shubham", 
   "role": "Android Developer", 
   "language": "Kotlin"
}

# modify
dict2["language"] = "Python"
print(dict2)
```

## Dictionary Methods
These are the tools you will use to deconstruct the hash map.

- keys(): Returns a view of all keys.
- values(): Returns a view of all values.
- items(): Returns a view of all (key, value) tuples. Extremely useful for iteration.
- pop(key): Removes the key and returns its value. Throws an error if the key isn't found (unless you provide a default: pop(key, default)).
- del my_dict[key]: Deletes the key-value pair, but doesn't return the value.

```python
dict3 = {
   30 : "Shubham",
   40 : "Prism",
   16 : "Python",
   17 : "Java",
   18 : "Kotlin"
}

# keys() - returns views of all keys
print(dict3.keys())

# values() - returns views of all values
print(dict3.values())

# items() - returns views of all key-value pairs as tuples
print(dict3.items())

# pop() - removes a key and returns its value
removed_value = dict3.pop(16)  # Removes key 16 and returns "Python"
print(removed_value)

# popitem() - removes and returns an arbitrary key-value pair as a tuple
removed_pair = dict3.popitem() # Removes & returns an arbitrary key-value pair
print(removed_pair)

# delete a key-value pair using del
del dict3[17]  # Deletes the key 17 and its associated value
print(dict3)

# clear() - removes all key-value pairs from the dictionary
dict3.clear()  # Clears all key-value pairs, leaving an empty dictionary
print(dict3)
```

## Iterating Over Dictionaries
When writing algorithms, you almost always need both the key and the value simultaneously.

```python
dict4 = {
   1 : "ant",
   2 : "bat",
   3 : "cat",
   4 : "dog",
   5 : "egg"
}

# The standard, Pythonic way to traverse a dictionary
for key, val in dict4.items():
   print(f"Key {key} = {val}")

# If you ONLY need keys (default behavior)
for key in dict4:
   print(key)

# If you ONLY need values
for val in dict4.values():
   print(val)
```

## Nested Dictionaries
Dictionaries can hold other dictionaries as values. This is how you build complex, heavily queried systems like Tries (Prefix Trees) or Weighted Graphs.

```python
# Adjacency list for a weighted graph
graph = {
    "A": {"B": 5, "C": 2},  # Node A connects to B (weight 5) and C (weight 2)
    "B": {"A": 5, "D": 1},
    "C": {"A": 2, "D": 8},
    "D": {"B": 1, "C": 8}
}

print(graph)
```

## Dictionary Comprehensions
Just like list comprehensions, these allow you to build dictionaries elegantly in a single line. This is a massive flex for writing clean, concise code.

```python
# Creating a freq map from an array
nums = [1, 2, 3, 4]
squares_map = {x: x**2 for x in nums}
# Result: {1: 1, 2: 4, 3: 9, 4: 16}
print(nums)
print(squares_map)
```

### The Core Concept: The $O(1)$ Lookup
- To truly understand why dictionaries are so fast, you have to visualize the hash function. 
- When you type my_dict["apple"] = 5, Python passes the string "apple" through a mathematical formula (a hash function) that spits out a memory index. 
- It then stores the value 5 exactly at that index.

- When you later ask for my_dict["apple"], it doesn't search through the dictionary. 
- It runs "apple" through the same math formula, gets the exact same memory index, and grabs the value instantly.
