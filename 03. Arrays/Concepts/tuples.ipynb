# Tuples

## Introduction to Tuples
- Tuples are ordered collections of items, just like lists.
- They maintain the exact order of the data you put into them, and they allow duplicate values.
- The critical difference is that once a tuple is created, its size and the memory references it holds cannot be altered.

## Creating Tuples
- Tuples are typically defined by parentheses ()
- But it is actually the comma that creates the tuple.

```python
# Standard creation
my_tuple = (4, 5)
print(my_tuple)

# Parentheses are technically optional (Packing)
colors = "red", "blue", "green"
print(colors)

# THE INTERVIEW TRAP: A single-element tuple
# This is an integer, NOT a tuple:
not_a_tuple = (5)   
print(not_a_tuple)

# This IS a tuple. The comma is mandatory:
real_tuple = (5,)
print(real_tuple)
```

## Accessing Elements
Because tuples are contiguous in memory, accessing any element via its index (or negative index) is an $O(1)$ operation.

```python
point = (10, 20, 30)
x = point[0]  # Returns 10 in O(1)
last = point[-1]  # Returns 30 in O(1)
print(x)
print(last)
```

## Tuple Operations
- We can use mathematical operators on tuple
- But we must remember that because tuples are immutable
- these operations create entirely new tuples in memory, running in $O(N)$ time.

```python
tup1 = (1, 2)
tup2 = (3, 4)
combined = tup1 + tup2  # Creates a NEW tuple (1, 2, 3, 4)
repeated = tup1 * 3  # Creates a NEW tuple (1, 2, 1, 2, 1, 2)
print(combined)
print(repeated)

print(type(tup1))  # <class 'tuple'>
print(type(combined))  # <class 'tuple'>
print(type(repeated))  # <class 'tuple'>
```

## The Immutable Nature (And Its Loophole)
- Immutability means we cannot add, remove, or reassign elements.
- if `tup = (1,2,3,4,5)`
- `tup[0] = 99` will throw a TypeError.

- However, there is a massive loophole that interviewers love to test:
- A tuple only locks the memory references, not the underlying objects.
- If a tuple contains a mutable object, like a list, we cannot replace the list, but we can modify the contents of the list itself.

```python
tup = (1, 2, 3)
# tup[0] = 10 -> This will raise a TypeError since tuples are immutable
print(tup)
print(tup[0])

my_tuple = (1, [2, 3], 4)
print(my_tuple)
my_tuple[1].append(5)  # Modifying the list inside the tuple
print(my_tuple)  # Output: (1, [2, 3, 5], 4)
```

## Tuple Methods
Because you cannot mutate a tuple, it lacks append(), remove(), pop(), or sort(). It only has two built-in methods:
- tup.count(val): Returns how many times a value appears in $O(N)$ time.
- tup.index(val): Returns the first index of a value in $O(N)$ time.

```python
tuple1 = (1, 2, 3)

# count
print(tuple1.count(2))  # Output: 1
# index
print(tuple1.index(3))  # Output: 2
```

## Packing and Unpacking
- Unpacking is one of Python's most elegant features.
- It allows us to assign the elements of a tuple to individual variables in a single line. 
- This operates in $O(1)$ time and is heavily used in algorithms like Breadth-First Search (BFS) to unpack coordinates.

```python
# Unpacking
node = (4, 5)
x, y = node
print(x)  # Output: 4
print(y)  # Output: 5

# The Swapping Trick (No temporary variable needed!)
a = 10
b = 20
a, b = b, a 
# Under the hood, this packs a tuple (b, a) and unpacks it into a, b
print(a)  # Output: 20
print(b)  # Output: 10
```

## Iterating Over Tuples
Iteration is identical to lists. It is highly optimized in Python.

```python
points = (0, 1, 0)
for val in points:
   print(val)

for i, val in enumerate(points):
   print(f"Index {i}: {val}")
```

## Nested Tuples
- We will use nested tuples heavily in graphing and matrix problems when we need to store visited states.
- A set cannot hold a list (because lists are mutable and unhashable), but a set can hold a tuple.

```python
# Storing grid coordinates we've already visited
visited = set()
current_node = (2, 3)  # Row 2, Col 3
visited.add(current_node)
print(visited)
```

## Tuple Comprehensions (The Generator Trap)
This is a tricky question interviewers use to see if we really know Python. There is no such thing as a tuple comprehension.

- If we use parentheses like a list comprehension, it does not create a tuple.
- It creates a Generator Object, which evaluates lazily to save memory.

```python
# List comprehension
my_list = [x**2 for x in range(3)]
print("My list: ", my_list)
print("Is it a list? ", type(my_list))
# Result: [0, 1, 4]

# The Trap: "Tuple Comprehension"
my_tup = (x**2 for x in range(3))
print("My tuple: ", my_tup)
print("But is it really a tuple? ", type(my_tup))
# Result: <generator object <genexpr> at 0x...>

# The Fix: If you actually need a tuple, you must pass the generator into tuple()
real_tuple = tuple(x**2 for x in range(3))
# Result: (0, 1, 4)
```
