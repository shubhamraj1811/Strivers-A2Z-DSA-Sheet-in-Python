# SETS

## What are sets?
- Sets are a built-in data type in Python used to store collections of unique items.
- They are unordered, meaning that the elements do not follow a specific order.
- They do not allow duplicate elements.
- Useful for membership tests, eliminating duplicate entries
- And performing mathematical set operations like
  - union
  - intersection
  - difference and symmetric difference

## Creating Sets
- Sets only store unique elements.
- If you pass duplicates during creation, Python instantly filters them out.

```python
my_set = {1,2,3,4,6,7,5,8,9,0}
print(my_set)
print(type(my_set))

# filters out duplicates
my_set2 = {1,2,3,4,4,4,4,5}
print(my_set2)

# empty set
my_empty_set = set()
print(my_empty_set)
```

```python
my_set3 = {2,1,3,4,5}
print(my_set3)

my_list = [2,2,1,1,3,3]
print(my_list)

set_from_list = set(my_list)
print(set_from_list)
```

## Set Operations
- `add(val)`: Adds an element in $O(1)$ average time.
- `remove(val)`: Removes an element in $O(1)$ time, but throws a KeyError if the value doesn't exist.
- `discard(val)`: Removes an element in $O(1)$ time and gracefully does nothing if the value isn't there (safer for interviews).
- `in` operator: (membership)
- if `val in my_set`: 
  - takes $O(1)$ time. 
  - Doing this in a list takes $O(N)$ time
  - this is the most common optimization you will make in brute-force algorithms.

```python
set_op = {1,2,3,4,5}

# adding an element to a set
set_op.add(6)
print(set_op)

# removing an element from a set
set_op.remove(3)
print(set_op)

# removing an element that does not exist in the set will raise a KeyError
# set_op.remove(10)
# print(set_op)

# dicard 
set_op.discard(1)
print(set_op)

# discard an element that does not exist in the set will not raise an error
set_op.discard(10)
print(set_op)

# in operator
print(2 in set_op)
print(10 in set_op)
print(10 not in set_op)
print(4 not in set_op)
```

- `pop` : removes an arbitrary element
- `clear all` : clears a set

```python
my_set4 = {1,2,3,4,5}
remove1 = my_set4.pop()
print(remove1)
print(my_set4)

# clear a set
my_set4.clear()
print(my_set4)
```

```python
# mathematical set operations

set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

## union
union_set = set1.union(set2)
print(union_set)

## intersection
intersection_set = set1.intersection(set2)
print(intersection_set)

## difference
difference_set = set1.difference(set2)
print(difference_set)

## symmetric difference
sym_diff_set = set1.symmetric_difference(set2)
print(sym_diff_set)
```
